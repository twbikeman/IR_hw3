```python
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import re
import os

stopWord = ['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', 'your', 'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', 'she', 'her', 'hers', 'herself', 'it', 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves', 'what', 'which', 'who', 'whom', 'this', 'that', 'these', 'those', 'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'having', 'do', 'does', 'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until', 'while', 'of', 'at', 'by', 'for', 'with', 'about', 'against', 'between', 'into', 'through', 'during', 'before', 'after', 'above', 'below', 'to', 'from', 'up', 'down', 'in', 'out', 'on', 'off', 'over', 'under', 'again', 'further', 'then', 'once', 'here', 'there', 'when', 'where', 'why', 'how', 'all', 'any', 'both', 'each', 'few', 'more', 'most', 'other', 'some', 'such', 'no', 'nor', 'not', 'only', 'own', 'same', 'so', 'than', 'too', 'very', 's', 't', 'can', 'will', 'just', 'don', 'should', 'now']
```

```python
entries = os.listdir('./')
p = re.compile('.*sgm')
fileList = []
for ent in entries:
    if p.match(ent):
        fileList.append(p.match(ent).group(0))
```

```python
with open("reut2-003.sgm", "r", encoding="utf-8") as fp:
    data = fp.read()
```

```python
soup = BeautifulSoup(data, "lxml")
```

```python
content = soup.select("REUTERS[TOPICS=YES][LEWISSPLIT=TRAIN] > TEXT")
contentNumber =len(content)
print(contentNumber)
regEx = re.compile('\\W+')
trainContent = []
for k in range(contentNumber):
    for i in (soup.select("TEXT")[k].children): pass
    beforeConv =i.string
    afterConv = [i for i in regEx.split(beforeConv) if not i in stopWord and len(i) >= 3]
    trainContent.append(afterConv)
```

```python
trainClass = soup.select("REUTERS[TOPICS=YES][LEWISSPLIT=TRAIN] > TOPICS ")
print(len(trainClass))
trainClassList = []
for k in range(len(trainClass)):
    trainClassList.append([i.string for i in trainClass[k].children])
```

```python
count = 0
for i in trainClassList:
    count += len(i)
print(count)
```

```python
df = pd.DataFrame(columns=['content','class'])
```

```python
for i in range(len(trainClass)):
    for j in trainClassList[i]:
        df.loc[len(df.index)] = [trainContent[i], j]
```

```python
df
```

```python
def createClassList(dataset):
    classList = set([])
    for document in dataset.iterrows():
        classList = classList | set([document[1]['class']])
    return list(classList)
```

```python
classlist = createClassList(df)

```

```python
def createVocabList(dataSet):
    vocabSet = set([])
    for document in dataSet.iterrows():
        vocabSet = vocabSet | set(document[1]['content'])
    return list(vocabSet)
        
```

```python
vacablist = createVocabList(df)
```

```python
def setOfWords2Vec(vocabList, inputSet):
    returnVec = [0] * len(vocabList)
    for word in inputSet:
        if word in vocabList:
            returnVec[vocabList.index(word)] +=  1
        else: print("the word is not in my vocabulary!")
    return returnVec
```

```python
def createTrainMatrix(vocabList, df):
    trainMatrix = []
    for i in range(df.shape[0]):
        vec = setOfWords2Vec(vocabList, df.iloc[i]['content'])
        trainMatrix.append(vec)
    return trainMatrix
```

```python
trainmatrix = createTrainMatrix(vacablist, df)
```

```python
print(len(trainmatrix))
```

```python
def createTrainCategory(df):
    trainCategory = []
    for i in range(df.shape[0]):
        trainCategory.append(df.iloc[i]['class'])
    return trainCategory
```

```python
traincategory = createTrainCategory(df)
```

```python
def trainNB0(trainMatrix, trainCategory, classList):
    numTrainDocs = len(trainMatrix)
    numWords = len(trainMatrix[0])
    pClass = {}
    pNum = {}
    pDenom = {}
    pVec ={}
    for i in classList:
        pClass[i] = sum([1 for item in trainCategory if item == i]) / float(numTrainDocs)
        pNum[i] = np.ones(numWords)
        pDenom[i] = float(numWords)
    for i in range(numTrainDocs):
        pNum[trainCategory[i]] += trainMatrix[i]
        pDenom[trainCategory[i]] += np.sum(trainMatrix[i])
    for i in classList:
        pVec[i] = np.log(pNum[i] / pDenom[i])
    return pClass, pVec, pNum, pDenom
```

```python
pclass, pvec, pnum, pdenom = trainNB0(trainmatrix, traincategory, classlist)
```

```python
pclass
```

```python
def classifyNB(vecClassify, testVec, pClass, classList):
    p = {}
    for i in classList:
        p[i] = sum(vecClassify[i] * testVec) + np.log(pClass[i])
    pClassMax = max(p.keys(), key=(lambda k: p[k]))
    pClassMaxValue = p[pClassMax]  
    return pClassMax
```

```python
success = 0
fail = 0
for i in range(len(trainmatrix)):
    guess = classifyNB(pvec, trainmatrix[i], pclass, classlist)
    if guess == df.iloc[i]['class']:
        success += 1
    else: fail += 1
print("precison: {}".format(success / (success + fail)))
```
